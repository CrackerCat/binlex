```python
import json
from pprint import pprint
from hashlib import sha256
import pybinlex
```

```python
raw = pybinlex.Raw()
raw.read_file('raw/raw.x86', 0)
raw_sections = raw.get_sections()
pprint(raw_sections)
```

```python
decompiler = pybinlex.Decompiler()
decompiler.setup(pybinlex.cs_arch.CS_ARCH_X86, pybinlex.cs_mode.CS_MODE_32, 0)
decompiler.set_mode("raw:x86", 0)
decompiler.set_corpus("default", 0)
decompiler.set_file_sha256(sha256(raw_sections[0]['data']).hexdigest(), 0)
decompiler.decompile(raw_sections[0]['data'], raw_sections[0]['offset'], 0)
traits = decompiler.get_traits()
print(json.dumps(traits, indent=4))
```
